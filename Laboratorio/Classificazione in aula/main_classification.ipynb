```python
from sklearn.datasets import load_iris
from DataManagment    import DataManagment
from Classifier       import Classifier
from sklearn.metrics  import accuracy_score, confusion_matrix
from scipy.stats      import spearmanr
import numpy as np
```

```python
iris_obj = DataManagment(load_iris, 0.3)
```

# SVC

```python
params = {
    "kernel": ['linear', 'poly', 'rbf', 'sigmoid'],
    "C": [0.1, 0.5, 1, 5, 100, 500],
    "gamma": ["auto"],
    "degree": list(range(3,6))
}

svc_obj = Classifier("svc")
svc_obj.best_parameters_computing(params, "accuracy", iris_obj, cv=10)
```

```python
targ_vt_tr_pr = svc_obj.final_model.predict(iris_obj.ftx_mtx_tr)
targ_vt_ts_pr = svc_obj.final_model.predict(iris_obj.ftx_mtx_ts)
```

```python
accuracy_tr = accuracy_score(targ_vt_tr_pr, iris_obj.targ_vt_tr)
accuracy_ts = accuracy_score(targ_vt_ts_pr, iris_obj.targ_vt_ts)
```

```python
print("Train accuracy:", accuracy_tr)
print("Test accuracy:", accuracy_ts)
```

```python
svc_obj.confusion_matrix_computing(targ_vt_tr_pr, targ_vt_ts_pr, iris_obj)
```

# Bayesian model

```python
features_number = len(iris_obj.ftx_names)
```

```python
for i in range(0, features_number - 1):
    for j in range(i + 1, features_number):
        corr_v, _ = spearmanr(iris_obj.ftx_mtx[:, i], iris_obj.ftx_mtx[:,j])
        print(
            "Feature 1: ", iris_obj.ftx_names[i],
            " Feature 2: ", iris_obj.ftx_names[j],
            " Correlazione: ", corr_v
            )        
```

```python
iris_obj.ftx_mtx_tr = iris_obj.ftx_mtx_tr[:,:2]
iris_obj.ftx_mtx_ts = iris_obj.ftx_mtx_ts[:,:2]
```

```python
params = {
    "alpha": [0.000001, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1],
    "force_alpha": [False, True]
}
```

```python
cnb_obj = Classifier("cnb")
cnb_obj.best_parameters_computing(params, "accuracy", iris_obj, cv=10)
```

```python
targ_vt_tr_pr = cnb_obj.final_model.predict(iris_obj.ftx_mtx_tr)
targ_vt_ts_pr = cnb_obj.final_model.predict(iris_obj.ftx_mtx_ts)
```

```python
accuracy_tr = accuracy_score(targ_vt_tr_pr, iris_obj.targ_vt_tr)
accuracy_ts = accuracy_score(targ_vt_ts_pr, iris_obj.targ_vt_ts)
```

```python
print("Train accuracy:", accuracy_tr)
print("Test accuracy:", accuracy_ts)
```
