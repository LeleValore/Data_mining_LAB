#### A. FREQUENZA ELEMENTI STESSO BACKET

```python
basket = basket.groupby(['InvoiceNo', 'Description'])["Quantity"].sum()
```

#### B. UNSTACK

Crea una matrice quadrata dove le righe sono le transazioni e le colonne tutte le possibili descrizioni.
Siccome tutti gli item non stanno in tutti i carrelli, dove troviamo NA mettiamo 0 (indica freq 0)

```python
basket = basket.unstack().reset_index().fillna(0).set_index("InvoiceNo")
```

```python
basket
```

#### C. BASKET ENCODED

Creiamo una matrice quadrata come quella di sopra dove però, al posto della frequenza, mettiamo 1 per indicare se l'elemento è presente, 0 altrimenti.

```python
def encode_values(x):
    return 0 if x <= 0 else 1

basket_encoded = basket.applymap(lambda element: encode_values(element))
```

#### D. FILTRARE TRANSAZIONI CON UN SOLO ELEMENTO

```python
condizione = (basket_encoded > 0).sum(axis=1) >= 2
basket_encoded_filter = basket_encoded[condizione]
```

#### E. APRIORI

<figure><img src="Figure/apriori.png" width="600" height="400"></figure>

L'algoritmo apriori assume che qualsiasi <b>subset</b> di un <b>itemset frequente</b> deve essere <b>frequente</b>.
Pertanto, se consideriamo la <b>transazione</b> riportata sopra <b>(vino, patatine, pane)</b>, ed assumiamo che quest'ultima <b>riulti essere frequente</b>, <b>allora</b> anche la <b>transazione</b> composta da <b>(vino, pane)</b> sarà frequente. Viceversa, se consideriamo un <b>itemset non frequente</b>, qualsunque <b>superset</b> sarà a sua volta <b>non frequente</b>.

<b> NOZIONI:</b>
Per selezionare regole di associazione interessanti da uno scenario di business, si adottano diverse metriche:
  - <b>Supporto</b>: rapporto tra numero di transazioni dove l'item x compare / numero totale di transazioni
  - <b>Confidenza</b>: (x => y) la probabilità di comprare l'item y dato per certo di comprare x. Esso tiene conto della popolarità dell'elemento x.
  
    Es: <b>(vino, pane) -> (patatine)</b>: 
    
    <b>confidenza</b>((vino, pane) -> (patatine)) = <b>supporto</b>((vino, pane, patatine)) / <b>supporto</b>((vino, pane))
  
  -  <b>lift</b>: misura di associazione tra due elementi in un insieme di dati transazionali, spesso utilizzata per identificare relazioni di co-occorrenza tra gli elementi.
  
     Es: <b>(vino, pane) -> (patatine)</b>: 
    
     <b>lift</b>(vino, pane) -> (patatine)) = <b>confidenza</b>(vino, pane) -> (patatine)) / <b>supporto</b>((patatine))
     
     Se il <b>lift</b> è uguale a 1, significa che non c'è alcuna associazione tra (vino, pane) e (patatine).
     
     Se il <b>lift</b> è maggiore di 1, indica che c'è un'associazione positiva tra (vino, pane) e (patatine). Un valore più alto di lift suggerisce una maggiore forza di associazione.
     
     Se il <b>lift</b> è inferiore a 1, indica un'associazione negativa tra (vino, pane) e (patatine). In altre parole, l'acquisto di (vino, pane) tende a ridurre l'acquisto di (patatine).
     
<figure><img src="Figure/apriori pruning.png" width="800" height="600"></figure>

<b>LIBRERIA UTILIZZATA</b>

mlxtend

apriori(df, min_support=0.5, use_colnames=False, max_len=None, verbose=0, low_memory=False)

 - min_support: supporto minimo che deve avere l'itemset per essere considerato frequente
 - max_len: massima dimensione dell'itemset che viene creato. se non settato va a creare tutte le possibili dimensioni.
 - low_memory: settare a True solo se si lavora con grandi dataset

```python
# ESEMPIO:

#(JUMBO BAG RED RETROSPOT -> PARTY BUNTING)
total = len(basket)
# supporto(JUMBO BAG RED RETROSPOT)
jbrr  = basket["JUMBO BAG RED RETROSPOT"]
jbrr  = jbrr[jbrr > 0]
support_jbrr = len(jbrr)/total
print("supporto di jbrr: ", support_jbrr)

# supporto(PARTY BUNTING)
pb    = basket["PARTY BUNTING"]
pb    = pb[pb > 0]
support_pb = len(pb)/total
print("supporto di pb: ", support_pb)

# supporto(JUMBO BAG RED RETROSPOT, PARTY BUNTING)
jbrr_pb = basket[["JUMBO BAG RED RETROSPOT", "PARTY BUNTING"]]
jbrr_pb = jbrr_pb[(jbrr_pb["JUMBO BAG RED RETROSPOT"] > 0) & (jbrr_pb["PARTY BUNTING"] > 0)]
support_jbrr_pb = len(jbrr_pb) / total
print("supporto di jbrr_pb: ", support_jbrr_pb)

# confidenza(JUMBO BAG RED RETROSPOT, PARTY BUNTING)
confidenza_jbrr_pb = support_jbrr_pb/support_jbrr
print("confidenza di jbrr_pb: ", confidenza_jbrr_pb)

# lift(JUMBO BAG RED RETROSPOT, PARTY BUNTING)
lift_jbrr_pb  = confidenza_jbrr_pb/support_pb
print("lift di jbrr_pb: ", lift_jbrr_pb)
```

#### APPLICAZIONE ALGORITMO

```python
frequent_itemsets = apriori(
    basket_encoded_filter, 
    min_support=0.03, 
    use_colnames=True
).sort_values("support",ascending=False)
```

```python
frequent_itemsets
```

Dai risultati possiamo notare che “Portalampada bianco a forma di cuore” è l'item più frequente con supporto di 0.121358. Esso è stato comprato nel 12% delle transazioni.

#### E. APPLICAZIONE REGOLE DI ASSOCIAZIONE

Step finale è trovare regole con supporto, confidence, and lift.

```python
association_rules(frequent_itemsets, metric="lift", min_threshold=1).sort_values("lift",ascending=False).reset_index(drop=True)
```

Dal risultato ottenuto si nota che, “Roses Regency Teacup and Saucer” e “Green Regency Teacup and Saucer” hanno un alto valore di lift, e quindi una buona associazione tra i due prodotti. Dal supporto si evince che i due elementi sono situati (insieme) nel 3% delle transazioni.
